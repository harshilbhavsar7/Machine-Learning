```python
import tensorflow as tf

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model=load_model('rps2.h5')
```

```python
def classifier(path):
    raw_img=image.load_img(path,target_size=(150,150))#converting image to target size
    raw_img=image.img_to_array(raw_img)#convert image to array
    plt.imshow(raw_img)
    raw_img=np.expand_dims(raw_img,axis=0)#expand dimensions
    raw_img=raw_img/255
    prediction=model.predict_classes(raw_img)
    accuracy = model.predict(raw_img)
    plt.imshow(cv2.imread(path))
    print('Accuracy ',accuracy)
    
    print(prediction)
    if (prediction==0):
        print("It's a paper ")
    elif(prediction==1):
        print("It's a rock ")
    elif(prediction==2):
        print("It's a scissors")
```
